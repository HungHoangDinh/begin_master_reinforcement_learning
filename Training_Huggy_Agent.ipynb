```python
config = """behaviors:
  Huggy:
    trainer_type: ppo
    hyperparameters:
      batch_size: 2048
      buffer_size: 20480
      learning_rate: 0.0003
      beta: 0.005
      epsilon: 0.2
      lambd: 0.95
      num_epoch: 3
      learning_rate_schedule: linear
    network_settings:
      normalize: true
      hidden_units: 512
      num_layers: 3
      vis_encode_type: simple
    reward_signals:
      extrinsic:
        gamma: 0.995
        strength: 1.0
    checkpoint_interval: 200000
    keep_checkpoints: 15
    max_steps: 2e6
    time_horizon: 1000
    summary_freq: 50000
"""

with open("/content/ml-agents/config/ppo/Huggy.yaml", "w") as f:
    f.write(config)
```

```python
!mlagents-learn ./config/ppo/Huggy.yaml --env=./trained-envs-executables/linux/Huggy/Huggy --run-id="Huggy" --no-graphics
```

```python
from huggingface_hub import HfApi, upload_file
from huggingface_hub import create_repo

# 1. THÔNG TIN CỦA BẠN (SỬA Ở ĐÂY)
# ======================================================
username = "hunghd20012003"  # Ví dụ: "nguyenvana"
repo_name = "my_huggy"             # Tên muốn đặt cho model
# ======================================================

repo_id = f"{username}/{repo_name}"

# 2. Tạo Repo trên Hugging Face (nếu chưa có)
print(f"Đang tạo repo: {repo_id}...")
create_repo(repo_id, repo_type="model", exist_ok=True)

# 3. Upload file Model (.onnx)
# Mặc định ML-Agents lưu kết quả trong thư mục 'results'
# Tên file thường là tên bạn đặt trong run-id (ví dụ: Huggy.onnx)
# Bạn hãy kiểm tra cột bên trái Colab xem file .onnx nằm chính xác ở đâu nhé
model_path = "./results/Huggy/Huggy.onnx" # <-- Kiểm tra đường dẫn này

print(f"Đang upload model từ {model_path}...")

try:
    upload_file(
        path_or_fileobj=model_path,
        path_in_repo="", # Tên file khi lên trên mạng
        repo_id=repo_id,
        repo_type="model",
        commit_message="Push Agent from Deep RL Course Unit Bonus 1"
    )
    print("✅ Upload thành công! Chúc mừng bạn.")
    print(f"Link model của bạn: https://huggingface.co/{repo_id}")

except Exception as e:
    print("❌ Có lỗi xảy ra:", e)
    print("Hãy kiểm tra lại đường dẫn file model_path xem đúng chưa nhé!")

# 4. (Tùy chọn) Upload thêm file config (.yaml) để sau này dễ xem lại
try:
    upload_file(
        path_or_fileobj="config.yaml", # File cấu hình huấn luyện
        path_in_repo="config.yaml",
        repo_id=repo_id,
        repo_type="model"
    )
    print("✅ Đã upload file config.")
except:
    print("⚠️ Không tìm thấy file config.yaml, bỏ qua.")
```
